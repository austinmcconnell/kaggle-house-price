```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
df.describe()
```

```python
df.columns
```

## Prediction Target

```python
plt.figure(figsize=(8, 4.5))
sns.distplot(df['SalePrice'], kde = False, color = 'b', hist_kws={'alpha': 0.9});
```

## Numerical Features

```python
df_numerical = df.select_dtypes(['int64', 'float64'])
df_numerical.columns
```

```python
corr = df_numerical.corr()
plt.figure(figsize=(12, 12));
sns.heatmap(corr, vmax=1, square=True)
```

```python
corr['SalePrice'].sort_values(ascending=False)
```

```python
sns.regplot(x = 'OverallQual', y = 'SalePrice', data = df, color = 'Orange');
```

## Categorical Features

```python
df_categorical = df.select_dtypes(['object'])
df_categorical.columns
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'Neighborhood', y = 'SalePrice',  data = df)
xt = plt.xticks(rotation=45)
```

```python
g = sns.FacetGrid(df, col = 'YrSold', col_wrap = 3)
g.map(sns.boxplot, 'MoSold', 'SalePrice', palette='Set2', order = range(1, 13))\
.set(ylim = (0, 500000))
plt.tight_layout()
```

```python
sns.boxplot(x = 'BldgType', y = 'SalePrice', data = df)
```

```python
sns.factorplot('HeatingQC', 'SalePrice', hue = 'CentralAir', estimator = np.mean, data = df, size = 4.5, aspect = 1.4);
```
